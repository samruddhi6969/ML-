```python
# SVM

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
# create database
data = {
    'age': [22,25,30,35,40,28],
    'salary': [25000, 30000, 50000, 65000, 80000, 48000],
    'purchased': [0,0,1,1,1,0]
}
df = pd.DataFrame(data)
df
```

```python
# features and target

X = df[['age', 'salary']]
y = df['purchased']

# feature scalling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train SVM Model:

model = SVC(kernel ='linear')
model.fit(X_scaled,y)

# prediction 

new_customer = scaler.transform([[32,60000]])
prediction = model.predict(new_customer)

print("Prediction (1= Buy, 0= Not Buy):",prediction[0])

# visualization

plt.scatter(df['age'], df['salary'], c= df['purchased'])
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Customer Purchase Dataset")
plt.show()
```

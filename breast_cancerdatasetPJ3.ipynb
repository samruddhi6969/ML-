```python
df = df.drop_duplicates()
```

```python
total_nulls = df.isnull().sum().sum()
total_nulls
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
)
df.columns
```

```python
df.duplicated().sum()
```

```python
plt.figure()
sns.boxplot(y=df['radius_mean'])
plt.title("Box Plot of Radius Mean")
plt.ylabel("Radius Mean")
plt.show()
```

```python
# scatter plot

plt.figure()
plt.scatter(df['radius_mean'],df['area_mean'])
plt.xlabel("radius mean")
plt.ylabel("area mean")
plt.title("radius vs area")
plt.show()
```

```python
plt.figure()
plt.scatter(df['texture_mean'], df ['smoothness_mean'])
plt.xlabel("texture mean")
plt.ylabel("smoothness mean")
plt.title("texture vs smoothness")
plt.show()
```

```python
# diagnosis vs radius

plt.figure()
sns.boxplot(x=df['diagnosis'], y=df['radius_mean'])
plt.title("Diagnosis vs Radius Mean")
plt.xlabel("Diagnosis (M = Malignant, B = Benign)")
plt.ylabel("Radius Mean")
plt.show()
```

```python
# radius mean hist
# distribution of radius

plt.figure()
plt.hist(df['radius_mean'])
plt.title("Histogram of Radius Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
# area mean hist

plt.figure()
plt.hist(df['area_mean'])
plt.title("Histogram of Area Mean")
plt.xlabel("Area Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.histplot(df['radius_mean'])
plt.title("Radius Mean Distribution")
plt.show()
```

```python
cols = df.columns
chunk_size = 9   # 3 x 3 grid

for i in range(0, len(cols), chunk_size):
  df[cols[i:i+chunk_size]].hist(figsize=(12,8))
  plt.suptitle(f"Features {i} to {i+chunk_size}")
  plt.tight_layout()
  plt.show()
```

```python
plt.scatter(df['radius_mean'], df['texture_mean'])
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.title("Radius vs Texture")
plt.show()
```

```python
df.boxplot(column= 'radius_mean', by='diagnosis')
plt.title("Radius Mean by Diagnosis")
plt.suptitle("")
plt.xlabel("Diagnosis")
plt.ylabel("Radius mean")
plt.show()
```

```python
features = ['radius_mean', 'texture_mean', 'area_mean']

for i in range(len(features)):
    for j in range(i+1, len(features)):
        plt.figure()
        plt.scatter(df[features[i]], df[features[j]])
        plt.xlabel(features[i])
        plt.ylabel(features[j])
        plt.title(f"{features[i]} vs {features[j]}")
        plt.show()
```

```python
corr_target = df.corr()['diagnosis'].sort_values()

corr_target.plot(kind='bar')
plt.title("Feature Correlation with Diagnosis")
plt.show()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
corr_target = df.corr()['diagnosis'].sort_values()

corr_target.plot(kind = 'bar')
plt.title("Feature Correlation with Diagnosis")
plt.shoe()
```

```python
#  Correlation Heatmap
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation = 90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

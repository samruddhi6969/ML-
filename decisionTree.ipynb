```python
# decision tree
# import necessary liabrarire
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data = {
    'Income': [30,40,50,60,70,80],
    'Credit score': [600,650,700,720,750,800],
    'Loan Approved': [0,0,1,1,1,1]
}
df = pd.DataFrame(data)
df
```

```python
# define features and target: 

X = df[['Income', 'Credit score']]
y = df['Loan Approved']

# train the model:
model = DecisionTreeClassifier(criterion = 'gini')
model.fit(X,y)

# prediction
prediction = model.predict([[55,710]])
print("Loan Approval Prediction (1 = Yes, 0 = No):", prediction[0])

```

```python
# visualize tree
plt.figure(figsize = (10,6))
tree.plot_tree(model,
              feature_names =['Income', 'Credit score'],
               class_names = ['No', 'Yes'],
               filled = True)
plt.title("Decision Tree Visualization")
plt.show()
```

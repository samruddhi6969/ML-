```python
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBClassifier, plot_importance

# Dataset
data = {
'Age':[22,25,30,35,40,28,32,45,38,29],
'Salary':[25000,30000,50000,65000,80000,48000,60000,90000,72000,52000],
'Purchased':[0,0,1,1,1,0,1,1,1,0]
}

df = pd.DataFrame(data)
df

X = df[['Age','Salary']]
y = df['Purchased']

# Train model
model = XGBClassifier(
    n_estimators=50,
    max_depth=3,
    learning_rate=0.1,
    eval_metric='logloss'
)

model.fit(X,y)

# Plot importance
plot_importance(model)
plt.title("Feature Importance - XGBoost")
plt.show()
```

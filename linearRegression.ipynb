```python
# import necessary liabrarire
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Experience' :[ 1, 2, 3, 4, 5, 6, 7],
    'salary' : [ 30000, 35000, 50000, 60000, 65000, 70000, 80000]
}
df = pd.DataFrame(data)
df
```

```python
# define features and target: 

X = df[['Experience']]
y = df['salary']

# train the model:
model = LinearRegression()
model.fit(X,y)

# prediction

predicted_salary = model.predict([[8]])
print(predicted_salary[0])
```

```python
# visualisation

plt.scatter(X,y, label = "actualdata")

plt.plot(X,model.predict(X), label = "Regression Line") 
plt.xlabel("Experience (years)")
plt.ylabel("salary")
plt.tight_layout()
plt.show()

```
